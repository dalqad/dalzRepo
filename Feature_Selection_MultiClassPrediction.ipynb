#Import necessary libraries

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report
from sklearn.feature_selection import SelectKBest
```

#Reading data set and EDA

```python
df= pd.read_csv("cirrhosis.csv")
df.head()
df.info()
df.describe()
df.columns
```

```python
df["Stage"].unique()
```

#Checking for nulls and duplicates

```python
df.isna().sum()
```

```python
df.duplicated().sum()
```

```python
df=df.dropna()
```

```python
df.isna().sum()
```

#Label encoding

```python
df.info()
```

```python
label_encoder= LabelEncoder()
columns_to_encode= ['Status', 'Drug', 'Sex', 'Ascites', 'Hepatomegaly', 'Spiders', 'Edema']
for col in columns_to_encode:
 df[col]= label_encoder.fit_transform(df[col])
```

```python
df.info()
```

#Splitting X and y

```python
X= df.drop(columns= 'Stage')
y= df['Stage']
```

```python
X_train, X_test, y_train, y_test= train_test_split(X,y, test_size= 0.3, random_state= 42)
```

#Model performance before feature selection

```python
model= RandomForestClassifier()
model.fit(X_train, y_train)
y_predict= model.predict(X_test)
print(classification_report(y_predict, y_test))
```

#Model performance after feature selection

```python
selector = SelectKBest( k=5)
selector.fit(X_train, y_train)
```

```python
original_feature_names = X_train.columns
selected_features = original_feature_names[selector.get_support()]
print(selected_features)
```

```python
model.fit(X_train, y_train)
y_predict= model.predict(X_test)
print(classification_report(y_predict, y_test))
```

Accuracy is slightly better as their were irrelevant features
